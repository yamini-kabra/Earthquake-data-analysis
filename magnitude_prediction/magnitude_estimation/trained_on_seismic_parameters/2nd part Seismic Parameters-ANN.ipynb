```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:/Users/HP/Downloads/2nd_part.csv")
n=60
df
```

```python
Time=[]
#pd.to_datetime(df.time[0])-pd.to_datetime(df.time[100])
for i in range(len(df)-(n-1)):
    Time.append((pd.to_datetime(df.time[i])-pd.to_datetime(df.time[i+(n-1)])).days)
    print(Time[i])
print(len(Time))
Time
```

```python
M_mean=[]
sum=0
for i in range(n):
    sum=sum+df.mag[i]
M_mean.append(sum)
for i in range(len(df)-n):
    sum=sum-df.mag[i]+df.mag[i+n]
    M_mean.append(sum)
print(len(M_mean))
M_mean
```

```python
E_mean=[]
E=0
for i in range(n):
    E=E+(10**(11.8+1.5*df.mag[i]))**0.5
E_mean.append(E/Time[0])
for i in range(len(df)-n):
    E=E-(10**(11.8+1.5*df.mag[i]))**0.5+(10**(11.8+1.5*df.mag[i+n]))**0.5
    E_mean.append(E/Time[i+1])
    #print(Time[i+1].days)
print(len(E_mean))
E_mean
```

```python
import math
b=[]
a=[]
for i in range(len(df)-(n-1)):
    MN1=0
    MN2=0
    Mi=0
    Mis=0
    for j in range(i,i+n):
        cnt=0
        for k in range(i,i+n):
            if df.mag[j]>=df.mag[k]:
                cnt=cnt+1
        Mi=Mi+df.mag[j]
        Mis=Mis+(df.mag[j]**2)
        MN1=MN1+df.mag[j]*math.log10(cnt)
        MN2=MN2+math.log10(cnt)
    #print(Mi**2,Mis)
    y=(n*MN1-Mi*MN2)/(Mi**2-n*Mis)
    b.append(y*-1)
    x=(MN2+y*df.mag[i])/n
    a.append(x)
    #msd.append((MN2-(x+y*df.mag[i]))**2/99)
    #print(i,y)
print(len(b))
print(b)
print(a)
#print(msd)
```

```python
msd=[]
for i in range(len(df)-(n-1)):
    p=0
    for j in range(i,i+n):
        cnt=0
        for k in range(i,i+n):
            if df.mag[j]>=df.mag[k]:
                cnt=cnt+1
        Ni=math.log10(cnt)
    p=p+(Ni-(a[i]-b[i]*df.mag[i]))**2
    msd.append(p/(n-1))
print(len(msd))
msd
```

```python
MD=[]
for i in range(len(df)-(n-1)):
    mm=0
    for j in range(i,i+n):
        if df.mag[j]>mm:
            mm=df.mag[j]
    MD.append(mm-a[i]/b[i])
MD
```

```python
lowest_mag=100
highest_mag=0
for i in range(len(df)):
    if df.mag[i]<lowest_mag:
        lowest_mag=df.mag[i]
    if df.mag[i]>highest_mag:
        highest_mag=df.mag[i]
lowest_mag,highest_mag
```

```python
#earthquakes above 5.5
import datetime
CT=[]
c=[]
for i in range(len(df)-(n-1)):
    any_reference_date = pd.to_datetime(df.time[i+n-1])
    ti=0
    si=0
    for j in range(i,i+n):
        if df.mag[j]>=5.5:
            cnt=cnt+1
            ti=ti+(pd.to_datetime(df.time[j])-any_reference_date).days
            si=si+((pd.to_datetime(df.time[j])-any_reference_date).days**2)
    if cnt==0:
        CT.append(0)
        c.append(0)
        continue
    x=ti/cnt
    y=si/cnt
    if x==0:
        c.append(si)
    else:
        c.append((si-x**2)/x)
    CT.append(x)
print(len(CT))
print(c)
CT
```

```python
x_pred_dict={'Time':Time,'M_mean':M_mean,'E_mean':E_mean,'b':b,'a':a,'msd':msd,'MD':MD,'y':y,'CT':CT,'c':c}
new_df = pd.DataFrame(x_pred_dict)
new_df
new_df = new_df.iloc[:-1 , :]
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras import models
from tensorflow.keras import layers
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
target = new_df['y']
# features = df1.drop(['mag','time','magType'],axis=1)
features = new_df[['Time','M_mean','E_mean','b','a','msd','MD','CT','c']]
target = target.to_numpy()
features = np.array(features)
train_features, test_features, train_target, test_target = train_test_split(features, 
                                                                            target, 
                                                                            test_size=0.33, 
                                                                            random_state=0)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_features[:, :] = sc.fit_transform(train_features[:, :])
test_features[:, :] = sc.transform(test_features[:, :])
```

```python
network = models.Sequential()

# Add fully connected layer with a ReLU activation function
network.add(layers.Dense(units=16, activation='tanh', input_shape=(train_features.shape[1],)))

# Add fully connected layer with a ReLU activation function
network.add(layers.Dense(units=4, activation='tanh'))

# Add fully connected layer with no activation function
network.add(layers.Dense(units=1))
```

```python
# Compile neural network
network.compile(loss='mse', # Mean squared error
                optimizer='RMSprop', # Optimization algorithm
                metrics=['mse','mae']) # Mean squared error
```

```python
network.fit(train_features, # Features
                      train_target, # Target vector
                      epochs=500, # Number of epochs
                      verbose=1, # No output
                      batch_size=100, # Number of observations per batch
                      validation_data=(test_features, test_target)) # Data for evaluation
```
