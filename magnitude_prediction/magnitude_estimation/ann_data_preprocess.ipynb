```python
import pandas as pd
import datetime
import math
```

```python
df = pd.read_csv("../../data/4th_part.csv")
df.size
df.head(5)
df1=df.copy()
```

```python
df.info()
```

```python
long_mod = 60
lat_mod = 24


# for ind in df.index:
#     long = math.floor(df['longitude'][ind])
#     lat = math.floor(df['latitude'][ind])
#     df1.loc[ind,'longitude'] = long % long_mod
#     df1.loc[ind,'latitude'] = lat % lat_mod

for ind in df.index:
    long = math.floor(df['longitude'][ind])
    lat = math.floor(df['latitude'][ind])
    if lat == 25 or lat == 26: 
        lat -= 1
    if lat == 27: 
        lat -= 2
    if long == 61 or long == 62:
        long -= 1
    if long ==  63:
        long -= 2
    df1.loc[ind,'longitude'] = long % long_mod
    df1.loc[ind,'latitude'] = lat % lat_mod
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras import models
from tensorflow.keras import layers
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
target = df1['mag']
# features = df1.drop(['mag','time','magType'],axis=1)
features = df1[['latitude','longitude','depth']]
target = target.to_numpy()
features = np.array(features)
train_features, test_features, train_target, test_target = train_test_split(features, 
                                                                            target, 
                                                                            test_size=0.33, 
                                                                            random_state=0)
train_features
```

```python
network = models.Sequential()

# Add fully connected layer with a ReLU activation function org= 3
network.add(layers.Dense(units=30, activation='tanh', input_shape=(train_features.shape[1],)))

# Add fully connected layer with a ReLU activation function org = 2
network.add(layers.Dense(units=20, activation='tanh'))

# Add fully connected layer with no activation function
network.add(layers.Dense(units=1))
```

```python
# Compile neural network
network.compile(loss='mse', # Mean squared error
                optimizer='RMSprop', # Optimization algorithm
                metrics=['mse','accuracy']) # Mean squared error
```

```python
network.fit(train_features, # Features
                      train_target, # Target vector
                      epochs=500, # Number of epochs
                      verbose=1, # No output
                      batch_size=100, # Number of observations per batch
                      validation_data=(test_features, test_target)) # Data for evaluation
```
