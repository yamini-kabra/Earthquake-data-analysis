```python
import pandas as pd
import datetime
import math
```

```python
df = pd.read_csv("../../data/1st_part.csv")
df.size
df.head(5)
df1=df.copy()
```

```python
long_mod = 60
lat_mod = 24


for ind in df.index:
    long = math.floor(df['longitude'][ind])
    lat = math.floor(df['latitude'][ind])
    df1.loc[ind,'longitude'] = long % long_mod
    df1.loc[ind,'latitude'] = lat % lat_mod
```
