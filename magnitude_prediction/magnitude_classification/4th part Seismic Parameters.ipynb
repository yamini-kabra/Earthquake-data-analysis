```python
import pandas as pd
import numpy as np
df = pd.read_csv("../../data/4th_part.csv")
n=75
from ipynb.fs.full.Seismic_Parameters import time,m_mean,e_mean,a_and_b,Msd,md,low_and_high,ct_and_c,yyy,Msd1
Time = time(df,n)
M_mean = m_mean(df,n)
E_mean = e_mean(df,n)
a,b = a_and_b(df,n)
msd = Msd1(df,n,a,b)
MD = md(df,n)
lowest_mag,highest_mag = low_and_high(df,n)
ct,c = ct_and_c(df,n)
y = yyy(df,n)
```

```python
Time=[]
#pd.to_datetime(df.time[0])-pd.to_datetime(df.time[100])
for i in range(len(df)-(n-1)):
    Time.append((pd.to_datetime(df.time[i])-pd.to_datetime(df.time[i+(n-1)])).days)
    print(Time[i])
print(len(Time))
Time
```

```python
M_mean=[]
sum=0
for i in range(n):
    sum=sum+df.mag[i]
M_mean.append(sum)
for i in range(len(df)-n):
    sum=sum-df.mag[i]+df.mag[i+n]
    M_mean.append(sum)
print(len(M_mean))
M_mean
```

```python
E_mean=[]
E=0
for i in range(n):
    E=E+(10**(11.8+1.5*df.mag[i]))**0.5
E_mean.append(E/Time[0])
for i in range(len(df)-n):
    E=E-(10**(11.8+1.5*df.mag[i]))**0.5+(10**(11.8+1.5*df.mag[i+n]))**0.5
    E_mean.append(E/Time[i+1])
    #print(Time[i+1].days)
print(len(E_mean))
E_mean
```

```python
import math
b=[]
a=[]
for i in range(len(df)-(n-1)):
    MN1=0
    MN2=0
    Mi=0
    Mis=0
    for j in range(i,i+n):
        cnt=0
        for k in range(i,i+n):
            if df.mag[j]>=df.mag[k]:
                cnt=cnt+1
        Mi=Mi+df.mag[j]
        Mis=Mis+(df.mag[j]**2)
        MN1=MN1+df.mag[j]*math.log10(cnt)
        MN2=MN2+math.log10(cnt)
    #print(Mi**2,Mis)
    y=(n*MN1-Mi*MN2)/(Mi**2-n*Mis)
    b.append(y*-1)
    x=(MN2+y*df.mag[i])/n
    a.append(x)
    #msd.append((MN2-(x+y*df.mag[i]))**2/99)
    #print(i,y)
print(len(b))
print(b)
print(a)
#print(msd)
```

```python
msd=[]
for i in range(len(df)-(n-1)):
    p=0
    for j in range(i,i+n):
        cnt=0
        for k in range(i,i+n):
            if df.mag[j]>=df.mag[k]:
                cnt=cnt+1
        Ni=math.log10(cnt)
    p=p+(Ni-(a[i]-b[i]*df.mag[i]))**2
    msd.append(p/(n-1))
print(len(msd))
msd
```

```python
MD=[]
for i in range(len(df)-(n-1)):
    mm=0
    for j in range(i,i+n):
        if df.mag[j]>mm:
            mm=df.mag[j]
    MD.append(mm-a[i]/b[i])
MD
```

```python
lowest_mag=100
highest_mag=0
for i in range(len(df)):
    if df.mag[i]<lowest_mag:
        lowest_mag=df.mag[i]
    if df.mag[i]>highest_mag:
        highest_mag=df.mag[i]
lowest_mag,highest_mag
```

```python
#earthquakes above 5.5
import datetime
CT=[]
for i in range(len(df)-(n-1)):
    cnt=0
    ti
    for j in range(i,i+n):
        if df.mag[j]>=5.5:
            cnt=cnt+1
            if ti:
                ti=ti+(datetime.datetime.now()-pd.to_datetime(df.time[j]))
            else:
                ti=(datetime.datetime.now()-pd.to_datetime(df.time[j]))           
    CT.append(ti/cnt)
print(len(CT))
CT
```

```python
y=[]
for i in range(len(df)-(n-1)):
    f1=0
    f2=0
    for j in range(i,i+n):
        if df.mag[i]>=5.8:
            f1=f1+1
        elif df.mag[i]>4.4:
            f2=f2+1
    if f1!=0:
        y.append(2)
    elif f2!=0:
        y.append(1)
    else:
        y.append(0)
print(y)
```

```python
x_pred_dict={'Time':Time,'M_mean':M_mean,'E_mean':E_mean,'b':b,'a':a,'msd':msd,'MD':MD,'y':y}
new_df = pd.DataFrame(x_pred_dict)
new_df
```

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = new_df
X = dataset.iloc[:, [0,1,2,3,4,5,6]].values
y = dataset.iloc[:, 7].values
print(X)
print(y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Fitting SVM to the Training set
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('\nAccuracy: {:.2f}\n'.format(accuracy_score(y_test, y_pred)))

print('Micro Precision: {:.2f}'.format(precision_score(y_test, y_pred, average='micro')))
print('Micro Recall: {:.2f}'.format(recall_score(y_test, y_pred, average='micro')))
print('Micro F1-score: {:.2f}\n'.format(f1_score(y_test, y_pred, average='micro')))

print('Macro Precision: {:.2f}'.format(precision_score(y_test, y_pred, average='macro')))
print('Macro Recall: {:.2f}'.format(recall_score(y_test, y_pred, average='macro')))
print('Macro F1-score: {:.2f}\n'.format(f1_score(y_test, y_pred, average='macro')))

print('Weighted Precision: {:.2f}'.format(precision_score(y_test, y_pred, average='weighted')))
print('Weighted Recall: {:.2f}'.format(recall_score(y_test, y_pred, average='weighted')))
print('Weighted F1-score: {:.2f}'.format(f1_score(y_test, y_pred, average='weighted')))
```
